```python
import datetime as dt
import pandas as pd
```

```python
seat_lst = []
```

```python
for i in range(1,5):
    start_seat = 40*i
    for j in range(11):
        seat = f'M{(start_seat+j):03d}'
        seat_lst.append(seat)
print(seat_lst)
```

```python
hour = ['00','01','02','03','04','05','06','07','08','17','18','19','20','21','22','23']
minute = ['00','29','30','59']
day = ['day_1','day_2','day_3','day_4','day_5']
time_lst = []
```

```python
for d in day:
    for h in hour:
        time_lst.append(f'{d}_{h}_{minute[1]}')
        time_lst.append(f'{d}_{h}_{minute[3]}')
print(time_lst)
#day_4_20_30
```

```python
header = ['id','seat_num']
for t in time_lst:
    header.append(t)
print(header)
```
